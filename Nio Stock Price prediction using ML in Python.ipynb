```python
# To get stock data connecting with API
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# To download data
Nio = yf.Ticker ('Nio')
```

```python
# To download recent data along with old data

Nio = Nio.history(period = "max")
Nio
```

```python
# For indexing
Nio.index
```

# Data Cleaning: Remove columns Dividends and Stock Splits

```python
# Will do some cleaning. We do not require the columns: Dividends and Stock Splits
del Nio["Dividends"]
del Nio["Stock Splits"]
Nio
```

```python
# To predict the stock price whether the price will go up or down we will use ML
Nio['Tomorrow'] = Nio['Close'].shift(-1)
Nio
```

```python
# Based on Tomorrow's price, we will set up a target and the target, we are going to predict using machine learning
Nio["Target"]= (Nio["Tomorrow"]>Nio["Close"]).astype(int)
Nio
```

```python
# We will take or consider data from 1st Jan 2020 to till 11th Jan, 2024
Nio=Nio.loc["2000-01-01":].copy()
Nio
```

# Machine Learning Model: RandomForestClassifier

```python
# We will set up our data by using machine learning model. Here we are going to use RandomForestClassifier Model.
from sklearn.ensemble import RandomForestClassifier

model=RandomForestClassifier(n_estimators=100, min_samples_split=100, random_state=1)

train = Nio.iloc[:-100]
test = Nio.iloc[-100:]

predictors=["Close", "Volume", "Open", "High", "Low"]    # List of predictors
model.fit(train[predictors], train["Target"])
```

```python
# To measure the accuracy of the model

from sklearn.metrics import precision_score

preds= model.predict(test[predictors])
import pandas as pd
preds = pd.Series(preds,index=test.index)
preds
```

```python
# To check the presision score using the actual target and the predictive target


precision_score(test["Target"], preds)

# the presicion score is 51%.
```

```python
# We will plot our predictions. To do that we will combine actual values with predicted values.
# For doing this we will use concate function.
combined=pd.concat([test["Target"], preds], axis=1)
```

```python
# We will plot it 

combined.plot()

# The orange line (0) is our prediction and the blue line is what actually happened.
# It shows that most of our prediction is that the price will go up
```

# Building a back testing system (To test out algorithm)

```python
#First we will create a prediction function

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
# Now we will create a backtest function

def backtest(data, model, predictors, start=500, step=250):
    all_predictions = []
    
    for i in range (start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
      
    return pd.concat(all_predictions)
        
```

```python
predictions = backtest(Nio, model, predictors)
```

```python
# Evaluating the erroe of our predictions 
# It will show how many days the market will go up vs. down
# Here In result, 0 represents that 585 days the market will be down and 1 represents that 258 days will be up.


predictions["Predictions"].value_counts()
```

```python
# We will check the precision score. The %age of time the market is up. 
# When we say the market will go up, our presicion score says that 46% of the time the market will go up.
# We are accurate 46%


precision_score(predictions["Target"], predictions["Predictions"])
```

```python
# We will check the %age of days the market actually was actually up vs. down
# It shows that in actual 52% days, the market was down and 47% days the market was up. 
# Almost same as our prediction model says

predictions["Target"].value_counts() / predictions.shape[0]
```

# Adding additional Predictors in our model: 

```python
# We will create variety of averages to determine whether the stock price will go up or down before buying the stock.
# We will create a horizon of rolling averages i.e. mean close price.
# Mean close price of last 2 days, last working trading week(5days), 30 days, 60 days, 250 days(one year), 500 days(2 years)
# Then we will also find the ratio with recent day's (today's) closing price.
# This will help us to make an idea (if the trend is down, it may turn up and if the trend is up, it may turn down) 

horizons = [2, 5, 30, 60, 250, 500, 750]
new_predictors = []

for horizon in horizons:
    rolling_averages = Nio.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    Nio[ratio_column] = Nio["Close"]/rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    Nio[trend_column] = Nio.shift(1).rolling(horizon).sum()["Target"]
    new_predictors += [ratio_column, trend_column]    
```

```python
Nio

# NaN means, there are no enough day to calculate the averages. 
```

```python
# NaN means, there are no enough day to calculate the averages. So we will drop the data which do not NaN

Nio=Nio.dropna()
Nio
```

```python
# Now we are going to modify or change something in our ML model. no_estimators as 200, min_samples_split as 50)

model=RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
```

```python
# Here in preds we will add proba (probability) to predict.
# Some conditions like if preds>=60%, then the market will be up and if preds<60%, the market will be down.

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds>=.6] = 1
    preds[preds<.6] =0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
predictions = backtest(Nio, model, new_predictors)
```
